```python
# --- 1. EDA & PREPROCESSING ---
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Veriyi Yükle
# Not: Bu link, projenin GitHub reposundaki veri setine işaret eder.
url = "https://raw.githubusercontent.com/KullaniciAdiniz/Metabolic-SERS-Diagnosis/main/metabolic_scores_final.csv"

try:
    df = pd.read_csv(url)
    # Gereksiz sütun temizliği
    if 'Sample' in df.columns:
        df = df.drop(columns=['Sample'])
    print("Veri Başarıyla Yüklendi! Boyut:", df.shape)
    display(df.head())
except:
    print("Veri okunamadı. Lütfen GitHub linkinin doğru olduğundan emin olun.")

# --- GÖRSELLEŞTİRME 1: Kutu Grafikleri (Metabolik Aktivite) ---
plt.figure(figsize=(15, 10))
features = ['Glikoliz', 'Lipid_Sentezi', 'Nukleotit', 'TCA_Dongusu']

for i, col in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(data=df, x='Cancer', y=col, palette="Set2")
    plt.title(f'{col} Aktivite Dağılımı')
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# --- GÖRSELLEŞTİRME 2: Korelasyon Matrisi ---
plt.figure(figsize=(8, 6))
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Metabolik Skorlar Arası İlişki")
plt.show()
```
